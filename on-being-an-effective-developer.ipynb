# On Being an Effective Developer
This notebook contains my notes from the *On Being an Effective Developer* talk by Raju Gandhi at *UberConf 2019*.

Raju said this talk contains mostly anecdotal personal discussion about improving individual and team effectiveness.

Raju offered the following definitions:
* *Productivity* - Getting the right things done
* *Efficiency* - Getting things done quickly
* *Effectiveness* - Productive + Efficient

### Personal Recommendations
He first gave some recommendations for increasing personal productivity

#### Say No
Raju said his only productivity tip in the talk is to "Say No". He said that context switches are expensive, and that it is therefore very important to limit work in progress. He pointed to Kanban as a methodology that has this as a cardinal rule. 

He recommends we don't answer questions other coworkers ask, and that we should redirect them to Google instead. His philosophy: "There is no such thing as a stupid question, but do your research first".

I personally disagree with that recommendation; I think it is an excellent way to burn your social capital with others. StackOverflow follows this mantra and it is not really a pleasant safe place to ask questions. I believe that building good relationships with others is more important than squeezing out every ounce of productivity.

However, I agreed with his limit work in progress recommendation and have found that to be true in my life.

#### Know Your Tools
Raju asked how many of us really know our tools. He gave some of the following examples:

* Git - `rebase -i`, `bisect`, and `worktree`
* Bash - `Ctrl-X, Ctrl-E` - Use to edit long commands more easily
* Editors (IntelliJ, Vim, etc.)

Some skills are worth learning well because we use them all the time, and they will be around for a lot longer than the current frameworks we are working with. 

The tools you use the most are the ones to focus on to become really effective. Raju recommended picking three tools to focus on and improve our knowledge of them.

He also asked how many of us are really good at Google searching, such as using advanced queries. He said that since so much of our work is stiching things together in ways others have also done, we should be very effective at finding and leveraging prior work.

#### Reduce Friction
The closer you bring things to yourself, the better off you will be. Raju have the example of Confluence: It always seems to be out-of-date. Things like READMEs in repositories or notes on a local laptop are usually more up-to-date. 

Raju asserts that the closer something is to us, the more likely we are to make it work.

He believes strongly in the power of plain-text notes (particularly markdown format in his case) as the best way to store information  because it can be transformed and used by many different tools. 

He offered several suggestions of good note apps:

* [nvAlt](https://brettterpstra.com/projects/nvalt/)
* [ResophNotes](http://resoph.com/ResophNotes/Welcome.html)
* [Notation](https://www.getnotation.com/)
* [Bear](https://bear.app/)

He said that Venkat (one of the other speakers at the conference) always automates opening up his presentations with a script. The script probably took a while to write, but he has used it countless times to easily set up a presentation.

### Mental Models
Raju said that the most effective people he knows are constantly refreshing their mental models. When a new tool version comes out, you should be learning the new features.

#### Reading Source Code
Raju recommended reading a lot of source code from other projects. He said he frequently reads the Spring source code. You can see how other people and groups do things.

Explore projects that are available in the world. Raju recommended the [Steal Like an Artist](https://www.amazon.com/Steal-Like-Artist-Things-Creative/dp/0761169253) book on this topic. He said it is important to read lots of code in lots of languages

#### Avoid NIH
Raju gave several examples of things we should be using but maybe aren't because of NIH syndrome:

* http://bash3boilerplate.sh/
* https://github.com/github/gitignore

#### Learn to Learn and Learn to Practice
He referenced *deliberate practice* and referenced the following books:

* [Peak](https://www.amazon.com/Peak-Secrets-New-Science-Expertise/dp/0544456238)
* [Deep Work](https://www.amazon.com/Deep-Work-Focused-Success-Distracted/dp/1455586692) 

I have read large parts of both those books and thought they were very applicable to being a programmer.

### Team Recommendations
Raju then gave some recommendations for increasing team efficiency

#### Communication
He talked about the importance of creating a shared context when communicating with others. Talking about UDP ports is usually not the correct shared context when talking with the VP of a company, but it is when talking with your team members.

Raju gave some good ways to help yourself develop the proper shared context:
* Creating presentations
* Writing about the topic

He also said the *two pizza rule* is important to help teams be efficient. Bigger teams means more lost connections

#### It's Not Personal
Raju said teams should try not to associate things like code reviews with personal attacks. Frankly I thought the sentiment was fine, but it lacked a lot of concrete details on how to effect that change in a team.

#### Workspace
Raju recommended having a workspace to *think*. He said "don't think in front of your computer. Think away from it" in a clean workspace with available design materials like whiteboards, pen/paper, etc.
